# **Creating and training Logestic regression**

```python
#  CREATING AND TRAINING LOGESTIC Regression model 
X = pd.DataFrame(data3 , columns = ["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"])
Y = data3.Outcome
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.25 , random_state = 0)
```

```python
logreg = LogisticRegression(solver = "liblinear")
```

```python
logreg.fit(X_train , Y_train)
Y_pred = logreg.predict(X_test)
```

```python
Y_pred
```

```python
"Accuracy: " , metrics.accuracy_score(Y_test , Y_pred)
```

```python
fpr , tpr ,_ = metrics.roc_curve(Y_test , Y_pred)
plt.plot(fpr , tpr , label = "data 1")
plt.legend(loc = 4)
plt.show
```

```python
Y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr , tpr,_ = metrics.roc_curve(Y_test , Y_pred_proba)
plt.plot(fpr , tpr , label = "data 1")
plt.legend(loc = 4)
plt.show()
```

```python
cm = confusion_matrix(Y , logreg.predict(X))
fig , ax = plt.subplots(figsize = (8,8))
ax.imshow(cm)
ax.grid(False)
ax.xaxis.set(ticks = (0,1) , ticklabels = ("predicted 0s" , "predicted 1s"))
ax.yaxis.set(ticks = (0,1), ticklabels = ("Actual 0s " , " Actual 1s"))
ax.set_ylim(1.5,-0.5)
for i in range(2):
  for j in range(2):
    ax.text(j,i,cm[i,j], ha = "center" , va = 'center' , color = "red")
plt.show()
```

```python
print(classification_report(Y , logreg.predict(X)))
```

# **creating and training KNN model**

```python
K = 3
CLF = KNeighborsClassifier(K)
CLF.fit(X_train,Y_train.ravel() )
Y_pred = CLF.predict(X_test)
```

```python
# Select Best value for K
K = 20
Acc = np.zeros((K))
for i in range(1 , K+1):
    CLF = KNeighborsClassifier(n_neighbors = i)
    CLF.fit(X_train,Y_train.ravel())
    Y_pred = CLF.predict(X_test)
    Acc[i-1] = metrics.accuracy_score(Y_test, Y_pred)
Acc
```

```python
# Show max & min
print(np.max(Acc))
print(np.min(Acc))
```

```python
traing_acc = []
test_acc = []
# try KNN for diffrent k nearset neighbor from 1 ta 50
neighbors_setting = range(1,20)

for n_neighbors in neighbors_setting:
    knn = KNeighborsClassifier(n_neighbors = n_neighbors)
    knn.fit(X_train,Y_train.ravel())
    traing_acc.append(knn.score(X_train,Y_train))
    test_acc.append(knn.score(X_test, Y_test))

plt.plot(neighbors_setting, traing_acc, label = "Accuracy of the training set")
plt.plot(neighbors_setting, test_acc, label = "Accuracy of the test set")
plt.ylabel("Accuracuy")
plt.xlabel("neighbors")
plt.grid()
plt.legend()
```
